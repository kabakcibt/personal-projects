```python
import random
import csv
import os

Dosya_Adi = "skorlar.csv"


def skor_kaydet(oyuncu, oyun, puan):
    dosya_var = os.path.exists(Dosya_Adi)

    with open(Dosya_Adi, "a", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)

        if not dosya_var:
            writer.writerow(["Oyuncu", "Oyun", "Puan"])

        writer.writerow([oyuncu, oyun, puan])


def skor_goster():
    try:
        with open(Dosya_Adi, "r", encoding="utf-8") as f:
            reader = csv.reader(f)
            satirlar = list(reader)

            if len(satirlar) <= 1:
                print("Henüz skor yok.")
                return

            print("\n===== SKORLAR =====")
            for s in satirlar:
                print(" | ".join(s))

    except FileNotFoundError:
        print("Dosya yok, olusturuluyor...")
        open(Dosya_Adi, "w").close()


def sayi_tahmin(oyuncu):
    hedef = random.randint(1, 100)
    hak = 7

    print("\n 1-100 Sayi Tahmin (7 hak)")

    while hak > 0:
        try:
            tahmin = int(input("Tahmin: "))
        except ValueError:
            print("Sadece sayi gir!")
            continue

        if tahmin == hedef:
            print("Bildin! +50")
            skor_kaydet(oyuncu, "Sayi Tahmin", 50)
            return
        elif tahmin < hedef:
            print("Daha büyük")
        else:
            print("Daha kücük")

        hak -= 1

    print(f"Bilemedin. Sayi: {hedef}")
    skor_kaydet(oyuncu, "Sayi Tahmin", 0)


def yazi_tura(oyuncu):
    secim = input("Yazi mi Tura mi (y/t): ").lower()

    if secim not in ["y", "t"]:
        print("Gecersiz secim")
        return

    sonuc = random.choice(["y", "t"])

    if secim == sonuc:
        print("Kazandin +20")
        skor_kaydet(oyuncu, "Yazi Tura", 20)
    else:
        print("Kaybettin")
        skor_kaydet(oyuncu, "Yazi Tura", 0)
```
