```python
department_counts = {}

with open("clean_logs.csv", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        department = parts[1]

        department_counts[department] = department_counts.get(department, 0) +1

        print("Bolum bazli hasta sayisi: ")
        for department, count in department_counts.items():
            print(department, "->", count)
```

```python
doctor_counts = {}

with open("clean_logs.csv", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        doctor = parts[3]

        doctor_counts[doctor] = doctor_counts.get(doctor, 0) +1
        print("Doktor bazli hasta sayisi:")
        for doctor, count in doctor_counts.items():
            print(doctor, "->", count)
```

```python
total_wait = 0
row_count = 0
with open("clean_logs.csv", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        waiting_time = int(parts[2])

        total_wait += waiting_time
        row_count += 1
        average_wait = total_wait / row_count

        print("Ortalama bekleme suresi: ",  round(average_wait, 2))
```

```python
long_wait = 0

with open("clean_logs.csv", "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(",")
        waiting_time = int(parts[2])
        if waiting_time > 50:
            long_wait += 1
            print("50 dakikadan fazla bekleyen hasta sayisi:", long_wait)
```
